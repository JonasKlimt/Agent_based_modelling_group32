```python
import itertools
from model import AdaptationModel
import pandas as pd
```

```python
# define model paramters

model_steps = 2 # maximum number of steps for each model run
model_runs = 5 # number of model runs for each combination of parameters

# subsidies levels
# information bias
```

```python


# Define fixed parameters
fixed_params = {
    "number_of_households": 50,
    "flood_map_choice": "harvey", 
    "network": "watts_strogatz", 
    "probability_of_network_connection": 0.4,
    "number_of_edges": 3,
    "number_of_nearest_neighbours": 5,    
}

# Define variable parameters
variable_params = {
    "subsidie_level": [0, 5000],
    "information_bias": [0, 0.1],
}

# Initialize a list to store the results
results = []

# Iterate over the variable parameters
for param_name, param_values in variable_params.items():
    for param_value in param_values:
        # Merge the fixed parameters and the current variable parameter
        params = {**fixed_params, param_name: param_value}
        
        # Run the model and collect the results
        for i in range(model_runs):  # number of model runs for each parameter value
            model = AdaptationModel(**params)
            for _ in range(model_steps):  # maximum number of steps for each model run
                model.step()
            model_data = model.datacollector.get_model_vars_dataframe().iloc[-1].to_dict()
            model_data.update(params)
            model_data["run"] = i
            results.append(model_data)

# Convert the results to a DataFrame
results_df = pd.DataFrame(results)

# Now `results_df` is a DataFrame that contains the results of all model runs
results_df
```
